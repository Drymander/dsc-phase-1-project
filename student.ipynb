```python
# tmdb_join['gross_profit'] = ('tmdb')
# https://pbpython.com/currency-cleanup.html
def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)

tmdb_join['production_budget'] = tmdb_join['production_budget'].apply(clean_currency).astype('float')
tmdb_join['domestic_gross'] = tmdb_join['domestic_gross'].apply(clean_currency).astype('float')
tmdb_join['worldwide_gross'] = tmdb_join['worldwide_gross'].apply(clean_currency).astype('float')
```

```python
tmdb_join['gross_profit'] = tmdb_join['worldwide_gross'] - tmdb_join['production_budget']
```
